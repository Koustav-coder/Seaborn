```python
flights=sns.load_dataset('flights')
```

```python
flights.pivot_table(index='month',columns='year',values='passengers',aggfunc='sum')
```

```python
pt=flights.pivot_table(index='month',columns='year',values='passengers',aggfunc='sum')
sns.heatmap(pt,cmap='autumn')
```

```python
sns.heatmap(pt,cmap='autumn',annot=True)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(pt,cmap='autumn',annot=True,linewidth=1)
```

```python
sns.clustermap(pt)
```

```python
#matplotlib
#2dplot
#pie
#Histogram
```

```python
new=flights.groupby('year').sum()['passengers'].reset_index()
```

```python
new
```

```python
new['something']=new['passengers']+500
```

```python
new
```

```python
plt.plot(new['year'],new['passengers'],color='red',marker='D',markersize='12',label='Passengers')
plt.plot(new['year'],new['something'],color='blue',marker='+',markersize='12',label='Something')
plt.xlabel('Year')
plt.ylabel('No. of passengers')
plt.title('US Airline Passenger Graph')
plt.legend()    #to add label
plt.show()
```

```python
#Histogram
plt.hist(tips['total_bill'],bins=50,color='black',label='something')
plt.legend()
plt.show()
```

```python
#pie Charts

new=tips.groupby('day').sum()['total_bill'].reset_index()
new
```

```python
plt.pie(new['total_bill'],labels=new['day'],autopct='%0.2f%%')
plt.show()
```
